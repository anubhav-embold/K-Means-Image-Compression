### Task 1: Importing Libraries

```python
from __future__ import print_function
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
plt.style.use("ggplot")

from skimage import io
from sklearn.cluster import KMeans

from ipywidgets import interact, interactive, fixed, interact_manual, IntSlider
import ipywidgets as widgets
```

```python
plt.rcParams['figure.figsize'] = (20, 12)
```

### Task 2: Data Preprocessing 
---

```python
# read the image & store it as a numpy array
img = io.imread('images/1-Saint-Basils-Cathedral.jpg')
# removing numbers from x & y axis
ax = plt.axes(xticks=[], yticks=[])
# adding a semicolon at end as we don't want to print the reference to this matplot object
ax.imshow(img);
# imshow() function in pyplot module of matplotlib library is used to display data as an image
```

```python
# shape of the array (height, width, depth) where depth is R,G,B color channels
img.shape
```

```python
# normalizing the pixel values to  re-scale between 0 & 1 instead of 0 & 255 unroll this array into 1 long vector for each of the 3 color channels
img_data = (img / 255.0).reshape(-1, 3)
img_data.shape
# now this is unrolled into one vector
```

```python
# MiniBatchKMeans helps in processing high-reosultion images in small batches
from sklearn.cluster import MiniBatchKMeans
```

### Task 3: K-means Image Compression with Interactive Controls
---

```python
img_dir = 'images/'
```

```python
@interact
def color_compression(image=os.listdir(img_dir), k=IntSlider(min=1,max=256,step=1,value=16,
                                                             continuous_update=False,
                                                             layout=dict(width='100%'))):
    
    input_img = io.imread(img_dir + image)
    img_data = (input_img / 255.0).reshape(-1, 3)
    
    kmeans = MiniBatchKMeans(k).fit(img_data)
    
    # from the returned fitted k-means object, we look at the cluster centres for which
    # we use k-means to predict the new color assignment on our original image
    k_colors = kmeans.cluster_centers_[kmeans.predict(img_data)]
    
    # After K-means has converged, load the large image into your program and replace each
    # of its pixels with the nearest of the centroid colors you found from the small image. 
    k_img = np.reshape(k_colors, (input_img.shape))
    # new image must have the same shape as the original image
    
    # (1,2) represents 1 row 2 columns
    fig, (ax1, ax2) = plt.subplots(1, 2)
    fig.suptitle('K-means Image Compression', fontsize=20)
    
    ax1.set_title('Compressed')
    ax1.set_xticks([])
    ax1.set_yticks([])
    ax1.imshow(k_img)
    
    ax2.set_title('Original (16,777,216 colors)')
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax2.imshow(input_img)
    
    plt.subplots_adjust(top=0.85)
    plt.show()
```
